```python
import PySimpleGUI as sg
from mysql import connector

sg.theme("GreenTan")

def sqlconnect(user,pas):
    con = connector.connect(
    host = "localhost",
    user = user,
    password = pas)
    return con

l1 = [ 
    [sg.Button("Connect to database")],
     [sg.Button("Enter User Query")],
    [sg.Button("Show Databases")],
      [sg.Text("Cancer database",size = (20,1), justification = "center")],
      [sg.Button("Types of cancer patients",size=(20,2))],
    [sg.Button("Benign cancer type",size=(20,2))],
    [sg.Button("Tumors with high radii",size=(20,2))],
    [sg.Button("Hard texture type cases",size=(20,2))],
[sg.Button("larger sized tumor cases",size=(20,2))]
     ]

l2 = [[sg.Text("Output",size = (10,15)),sg.Text("",size = (60,15),key = "out")]]

layout = [[sg.Column(l1),
           sg.VSeperator(),
           sg.Column(l2)]]

window = sg.Window("Dashboard",layout)

while True:
    event, values = window.read()
    
    if event in (sg.WIN_CLOSED, 'Exit'):
        break
    if event == sg.WIN_CLOSED or event == 'Exit':
        break
        
    if event:
        
        if event == "Connect to database":
            user = sg.popup_get_text("User name")
            password = sg.popup_get_text("password", password_char="*")
            sqlconnection = sqlconnect(user,password)
            window["out"].Update("Connected!")
            cur = sqlconnection.cursor()
            
        if event == "Enter User Query":
            q = sg.popup_get_text("enter your query")
            cur.execute(q)
                
            for x in cur:
                window["out"].Update(list(x))
        
        if event == "Show Databases":
            q= "show databases;"
            cur.execute(q)
            out = []
            for i in cur:
                out.append(list(i))
            
            window["out"].Update(out)
          
                
        if event == "Types of cancer patients":
            
            query1 = "use cancerdata;"
            query2 = "select diagnosis, count(diagnosis) from cancerdata1 group by diagnosis;"
            
            cur.execute(query1)
            cur.execute(query2)
            
            x = []
            for i in cur:
                x.append(list(i))
            
            output = "[diagnosis count] \n \n"
            for i in x:
                output = output + str(i) + "\n \n"
                
            window["out"].Update(output)
        
        if event == "Benign cancer type":
            
            query1 = "use cancerdata;"
            query2 = "select * from cancerdata1 inner join cancerdata2 on cancerdata1.id = cancerdata2.id where diagnosis = 'B';"
            
            cur.execute(query1)
            cur.execute(query2)
            
            x = []
            for i in cur:
                x.append(list(i))
            
            output = "[id diagnosis radius_mean texture_mean perimeter_mean area_mean smoothness_mean compactness_mean concavity_mean concave-point id symmetry_mean fractal_dimension_mean radius_se perimeter_se area_se smoothness_se compactness_se] \n \n"
            for i in x:
                output = output + str(i) + "\n \n"
                
            window["out"].Update(output)
        
        if event == "Tumors with high radii":
            
            query1 = "use cancerdata;"
            query2 = "select cancerdata1.id, cancerdata1.radius_mean, cancerdata1.diagnosis, cancerdata2.fractal_dimension_mean from cancerdata1 inner join cancerdata2 on cancerdata1.id = cancerdata2.id where radius_mean > 15;"
            
            cur.execute(query1)
            cur.execute(query2)
            
            x = []
            for i in cur:
                x.append(list(i))
            
            output = "[id radius_mean diagnosis fractal_dimension_mean] \n \n"
            for i in x:
                output = output + str(i) + "\n \n"
                
            window["out"].Update(output)
        
        if event == "Hard texture type cases":
            
            query1 = "use cancerdata;"
            query2 = "select cancerdata1.id, cancerdata1.diagnosis, cancerdata2.symmetry_mean from cancerdata1 inner join cancerdata2 on cancerdata1.id = cancerdata2.id where texture_mean > 20;"
            cur.execute(query1)
            cur.execute(query2)
            
            x = []
            for i in cur:
                x.append(list(i))
            
            output = "[id diagnosis symmetry_mean] \n \n"
            for i in x:
                output = output + str(i) + "\n \n"
                
            window["out"].Update(output)
        
        if event == "larger sized tumor cases":
            
            query1 = "use cancerdata;"
            query2 = "select cancerdata2.texture_se, cancerdata1.diagnosis,cancerdata1.area_mean from cancerdata1 inner join cancerdata2 on cancerdata1.id = cancerdata2.id where area_mean > 1000;"
            cur.execute(query1)
            cur.execute(query2)
            
            x = []
            for i in cur:
                x.append(list(i))
            
            output = "[texture_se diagnosis area_mean] \n \n"
            for i in x:
                output = output + str(i) + "\n \n"
                
            window["out"].Update(output)
        
            
window.close()
```
